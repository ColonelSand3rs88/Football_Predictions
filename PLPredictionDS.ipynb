```python
#next I will create predictors for our machine learning model
```

```python
games["home_formation_code"] = games["home_club_formation"].astype("category").cat.codes
#creates a column which sets the home formation type as an integer so the algorithm can use it for predictions
```

```python
games["away_formation_code"] = games["away_club_formation"].astype("category").cat.codes
#creates a column which sets the away formation type as an integer so the algorithm can use it for predictions
```

```python
games["home_club_code"] = games["home_club_id"].astype("category").cat.codes
#creates a column which sets the home formation type as an integer so the algorithm can use it for predictions
```

```python
games["away_club_code"] = games["away_club_id"].astype("category").cat.codes
#creates a column which sets the home formation type as an integer so the algorithm can use it for predictions
```

```python
games["target"] = np.where(games["home_club_goals"] > games["away_club_goals"],1,0)
#creates a conditonal column using numpys built-in where function to be used as a target for the prediction model if 1 home win, else 0
```

```python
games.dtypes
```

```python
from sklearn.ensemble import RandomForestClassifier
#imports model from sklearn that picks out non-linearities in the data
```

```python
rf = RandomForestClassifier(n_estimators=500, min_samples_split=100, random_state=1)
#n_estimators= number of individual decision trees to be trained 
#min_samples_split= number of samples we have in the decision trees
#random_state = 1 if we run mulitple times data will now stay the same
```

```python
train = games[games["date"] < '2019-01-01']
#this trains the model on the data before 2023
```

```python
test = games[games["date"] > '2019-01-01']
#this test the model on the data after 2023
```

```python
predictors = ["home_club_code","away_club_code","home_formation_code","away_formation_code"]
#selects the above columns to be used as predictors in our model
```

```python
rf.fit(train[predictors], train["target"])
#trains the random forest model with the above predictors
```

```python
preds = rf.predict(test[predictors])
#this passes in the test data and predictors
```

```python
from sklearn.metrics import accuracy_score
#this imports an accuracy score metric from sklearn and will allow us to show what percentage of the time the prediction was accurate
```

```python
acc = accuracy_score(test["target"], preds)
#this passes in our actuals and predictions
```

```python
acc
#this shows how often the prediction is accurate
```

```python
combined = pd.DataFrame(dict(actual=test["target"], prediction=preds))
#created a DataFrame to compare our actuals and predictions
```

```python
pd.crosstab(index=combined["actual"], columns=combined["prediction"])
#cross reference table to show how often prediction matched the actual
```

```python
grouped_games = games.groupby("home_club_name")
#create a DataFrame for each team for our data
```

```python
def rolling_averages(group, cols, new_cols):
    #adds to the precision of our model taking into account rolling average
        group = group.sort_values("date")
    #sorts by date
        rolling_stats = group[cols].rolling(3, closed="left").mean()
    #takes the last 3 values and takes the mean
        group[new_cols] = rolling_stats
        group = group.dropna(subset=new_cols)
    #drops any missing values so that we don't get errors where datasubset is too small
        return group
```

```python
cols = ["home_club_goals", "away_club_goals", "home_club_position", "away_club_position"]
new_cols = [f"{c}_rolling" for c in cols]
#takes the original column names and adds _rolling on the end to show they're rolling averages
```

```python
new_cols
```

```python
rolling_averages(group, cols, new_cols)
```

```python
games_rolling = games.groupby("home_club_name").apply(lambda x: rolling_averages(x, cols, new_cols))
#applies the rolling averages to our original dataset
```

```python
games_rolling
```

```python
games_rolling = games_rolling.droplevel('home_club_name')
#this drops the home team name index level that was applied as it is not needed
```

```python
games_rolling
```

```python
def make_predictions(data, predictors):
    train = data[data["date"] <'2019-01-01']
    test = data[data["date"] > '2019-01-01']
    rf.fit(train[predictors], train["target"])
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual=test["target"], predicted=preds), index=test.index)
    precision = precision_score(test["target"], preds)
    return combined, precision
    #the above combines our previous test and train set into one place making it easier to repeat the test following any changes
```

```python
combined, precision = make_predictions(games_rolling, predictors + new_cols)
#combines the original predictors with the new column to see how often the algorithm correctly predicts a home win
```
